### カラーバー
カラーバーは、ほとんどの2次元図につけるべきものである。matplotolibでは、かなり簡単につけることができるようになっているので必ずつけましょう。
その中でも図が**一つ**の時はかなり簡単で以下のようにやります。

```python
#Jupyter notebookのための設定
%matplotlib inline

#ライブラリの読み込み
import numpy as np
import matplotlib.pyplot as plt

#座標の設定
x = np.linspace(0,np.pi*2.,32)
y = np.linspace(0,np.pi*2.,32)

X, Y = np.meshgrid(x,y,indexing='ij')

z = sin(X)*cos(2*Y)

#コンター図作成はpcoloirmeshがおすすめです。
plt.pcolormesh(x,y,z,vmax=1.,vmin=-1.)
plt.colorbar()
```

図が二つ以上ある時は、以下のようにやります。

```python
#Jupyter notebookのための設定
%matplotlib inline

#ライブラリの読み込み
import numpy as np
import matplotlib.pyplot as plt

#座標の設定
x = np.linspace(0,np.pi*2.,32)
y = np.linspace(0,np.pi*2.,32)

X, Y = np.meshgrid(x,y,indexing='ij')

z0 = sin(X)*cos(2*Y)
z1 = 2.*exp(-X**2-Y**2)

#複数の図を並べる時は1次元の時と同じくインスタンスを作成
fig = plt.figure(num=1,figsize=(5,10))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

#colobarで使うので、pcolormeshの戻り値をim1, im2として保存しておく
im1 = ax1.pcolormesh(x,y,z0,vmax=1.,vmin=-1.)
im2 = ax2.pcolormesh(x,y,z1,vmax=2.,vmin=-2.)

#pcolormeshの戻り値を引数として、axesインスタンスをオプションとしてつけて
#カラーバーを作成します。オプションがないと、どちらのカラーバーも最後の図についてしまいます。
#カラーバーの戻り値も保存しておくと、あとでタイトルをつけることができます。
cl1 = fig.colorbar(im1,ax=ax1)
cl2 = fig.colorbar(im2,ax=ax2)

#カラーバーの図のメモリを大きくするのは以下
cl1.ax.tick_params(labelsize=15)
cl2.ax.tick_params(labelsize=15)

#タイトルをつける時は以下のようにする
#カラーバーのタイトルには単位などをつけると良いでしょう
cl1.ax.set_title("cm/s",fontsize=15,pad=10)
cl2.ax.set_title("cm/s",fontsize=15)
```
